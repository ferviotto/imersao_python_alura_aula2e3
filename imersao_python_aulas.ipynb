```python
import pandas as pd
import plotly.express as px
```

Aula 3 - Análise

```python
df_principal = df_principal[['Ativo','Data','Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo','Data':'data','Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct']/100
df_principal ['valor_inicial'] = df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['var_rs'] = (df_principal['valor_final']- df_principal['valor_inicial']) *df_principal ['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qntde_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['var_rs'].apply(lambda x:'Subiu' if x>0 else ('Desceu' if x<0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Resultado':'resultado','Ticker':'ticker','Nome':'nome','Segmento':'segmento','Idade (anos)':'idade_anos'}).copy()
df_principal
```

```python
df_principal['categoria_idade'] = df_principal['idade_anos'].apply(lambda x: 'Maior de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['var_rs'].max()

# Calculando o menor valor
menor = df_principal['var_rs'].min()

# Calculando a média
media = df_principal['var_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("segmento")["var_rs"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['var_rs'].sum().reset_index()

df_analise_saldo
```

```python
# Criando o gráfico de barras usando Plotly Express
fig = px.bar(df_analise_saldo, x='resultado', y='var_rs', text='var_rs', title='Variação Reais por Resultado')

# Atualizar a formatação dos números no eixo Y
fig.update_traces(texttemplate='$%{text:.2f}', textposition='outside')

# Mostrar o gráfico
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='segmento', values='var_rs', title='Variações Reais por Segmentos')
fig.show()
```

```python
segmentos_excluir = ['Petróleo e Gás', 'Mineração', 'Energia Elétrica', 'Bancário']
df_analise_segmento_filtrado = df_analise_segmento[~df_analise_segmento['segmento'].isin(segmentos_excluir)]

fig = px.pie(df_analise_segmento_filtrado, values='var_rs', names='segmento',
             title='Distribuição das Variações por Segmento')
fig.show()
```

```python
df_analise_faixa_etaria = df_principal.groupby('categoria_idade').agg({'var_rs': 'sum', 'ativo': 'count'}).reset_index()
df_analise_faixa_etaria
```

```python
df_analise_faixa_etaria.columns = ['analise_faixa_etaria', 'var_rs', 'qntde_empresas']
df_analise_faixa_etaria
```

```python
fig = px.bar(df_analise_faixa_etaria, x='analise_faixa_etaria', y='var_rs',
             text='var_rs', title='Variação de Receita por Faixa Etária')

fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()

```
